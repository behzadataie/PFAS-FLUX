# PFAS-FLUX technical note workflow

This notebook runs the same workflow as `run_pfas_flux.py` for the synthetic benchmark archive. It is intended as Supporting Information for the Hydrogeology Journal technical note.

```python
from pathlib import Path
from pfas_flux.io import read_archive, ensure_output_dirs, write_table
from pfas_flux.calculations import deterministic_workflow
from pfas_flux.uncertainty import MonteCarloOptions, monte_carlo_workflow
from pfas_flux.plots import plot_all

input_dir = Path('data/synthetic_case_archive')
output_dir = Path('outputs_notebook')
selected_group_id = 'G3'  # PFOS + PFHxS in the synthetic archive
paths = ensure_output_dirs(output_dir)
tables = read_archive(input_dir)
list(tables)
```

```python
det = deterministic_workflow(tables, selected_group_id=selected_group_id)
for name, df in det.items():
    write_table(df, paths['tables'], f'{name}.csv')
det['control_plane_group_summary'].query("group_id == @selected_group_id")
```

```python
options = MonteCarloOptions(n=10000, seed=20260703, selected_group_id=selected_group_id)
unc = monte_carlo_workflow(tables, options)
for name, df in unc.items():
    write_table(df, paths['tables'], f'{name}.csv')
unc['uncertainty_control_plane_group_summary'].query("group_id == @selected_group_id")
```

```python
figures = plot_all(det, unc, paths['figures'], selected_group_id=selected_group_id)
figures
```

## Caution

The receptor criterion in the synthetic archive is a placeholder. Replace it before any real application. The `F_comp` adjusted values are scenario calculations, not measured target-suite PFAS.
